```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('advertising.csv')
```

```python
data
```

```python
##whats the relationship between sales of tv and newspaper
```

```python
#missing values in the dataset; if so, in which column

missing_values = data.isnull().sum()
print('missing values:\n', missing_values)
```

```python
##amount of sales made for tv

tv_sales = data['TV'].value_counts()
print('tv_sales:\n' , tv_sales)
```

```python
##SALES DISTRIBUTION IN THE DATASET

plt.figure(figsize=(10,5))
sns.histplot(data['TV'], bins=5)
plt.title('TV distribution')
plt.xlabel('TV')

plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data['Newspaper'], bins=5)
plt.title('Newspaper distribution')
plt.xlabel('Newspaper')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data['Radio'], bins=5)
plt.title('Radio distribution')
plt.xlabel('Radio')
plt.ylabel('Frequency')
```

```python
data['Sales'].value_counts().plot(kind='bar');
```

```python
##distribution of the sales

plt.figure(figsize=(10,5))
sns.histplot(data['Sales'], bins=20)
plt.title('Sales distribution')
plt.xlabel('Sales')
plt.ylabel('Freqeuncy')
```

```python
##correlation and relationship
##correlation between Tv and SALES
sns.boxplot(x='Sales' , y='TV' , data=data)
plt.title('TV VS SALES')
plt.show()
```

```python
##CORRELATION BETWEEN SALES AND RADIO
sns.boxplot(x='Sales' , y='Radio' , data=data)
plt.title('SALES VS RADIO')
plt.show()
```

```python
##correlation between Newspaper and SALES
sns.boxplot(x='Sales' , y='Newspaper' , data=data)
plt.title('SALES VS NEWSPAPER')
plt.show()
```
